```python
#Calling the alcohol consumption dataset
file = r"C:\Users\zacha\Downloads\total-alcohol-consumption-per-capita-litres-of-pure-alcohol.csv"

df_Consumption_Per_Person = pd.read_csv(file)

# narrowing down the dataframe so that the only year recorded is the most recent, 2018
specific_year = 2018

# Filtering the rows
df_Consumption_Per_Person = df_Consumption_Per_Person[df_Consumption_Per_Person['Year'] == specific_year]

#Sorting the countries from most alcohol consumed to least alcohol consumed

df_Consumption_Per_Person_sorted = df_Consumption_Per_Person.sort_values(by='Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)', ascending=False)

#Reading the dataframe
df_Consumption_Per_Person_sorted
```

```python
file_No_Alcohol = r"C:\Users\zacha\Downloads\share-of-population-who-never-drink-alcohol.csv"
#Calling the dataset of the amount of people in a country who never drink alcohol
df_Never_Alcohol = pd.read_csv(file_No_Alcohol)

#Sorting the countries from who has the largest share of non-alcohol drinkers

df_Never_Alcohol_sorted = df_Never_Alcohol.sort_values(by='Indicator:Alcohol, abstainers lifetime (%) - Sex:Both sexes', ascending=False)

df_Never_Alcohol_sorted
```

```python
#Calling the "Happiness" dataset
file_Happiness = r"C:\Users\zacha\Downloads\happiness-cantril-ladder.csv"
df_Happiness = pd.read_csv(file_Happiness)

# narrowing down the dataframe so that the only year recorded is the most recent, 2022
specific_year = 2022

# Filtering the rows
df_Happiness = df_Happiness[df_Happiness['Year'] == specific_year]

#Sorting the countries starting with most self-reported happiness

df_Happiness_sorted = df_Happiness.sort_values(by='Cantril ladder score', ascending=False)

df_Happiness_sorted
```

```python
#Calling the tobacco smoking dataset

file_Smoke = r"C:\Users\zacha\Downloads\share-of-adults-who-smoke.csv"

df_Smoke = pd.read_csv(file_Smoke)

# Filtering the rows to only include data from the most recent year
specific_year = 2020
df_Smoke = df_Smoke[df_Smoke['Year'] == specific_year]

#Putting the smoking dataframe in order from countries who have the most active tobacco users to the least active tobacco users

df_Smoke_sorted = df_Smoke.sort_values(by='Prevalence of current tobacco use (% of adults)', ascending=False)

df_Smoke_sorted
```

```python
# Merging Alcohol Consumption dataframe with Smoking dataframe along three columns
df_Alcohol_Smoke = pd.merge(df_Consumption_Per_Person_sorted, df_Smoke_sorted, on=['Entity', 'Code'], how='inner')

# Merging the Alcohol and Smoke dataframe with the Happiness dataframe along the same three columns
df_Happiness_Alcohol_Smoke = pd.merge(df_Alcohol_Smoke, df_Happiness_sorted, on=['Entity', 'Code'], how='inner')

# Display the final dataframe
df_Happiness_Alcohol_Smoke
```
